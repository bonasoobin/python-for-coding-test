### 카드 정렬하기

[백준 링크](https://www.acmicpc.net/problem/1715) []

+ 정렬된 두 묶음의 숫자 카드가 있을 때 각 묶음의 카드의 수를 A,B라고 하면  
  보통 두 묶음을 합쳐서 하나로 만드는 데에는 A+B번의 비교를 해야함  
  (ex 20장의 카드 묶음과 30장의 카드 묶음을 합치려면 50번의 비교가 필요함)
+ 매우 많은 숫자 카드 묶음이 책상 위에 놓여있음
+ 두 묶음씩 골라 서로 합쳐나간다면, 고르는 순서에 따라서 비교 횟수가 매우 달라짐
+ 예를 들어 10장, 20장, 40장의 묶음이 있다면 10장과 20장을 합친 뒤, 합친 30장의 묶음과 40장을 합치면 (10+20) + (30+40) = 100  
  이지만, 10장과 40장을 합친 뒤, 합친 50장 묶음과 20장을 합친다면 (10+40)+(50+20) = 120이므로 덜 효율적임

+ N개의 숫자 카드 묶음의 각각의 크기가 주어질 때, 최소한 몇 번의 비교가 필요한지를 구하는 프로그램을 작성하기

**입력 조건**
+ 첫째 줄에 N이 주어짐(1<=N<=100,000) 이어서 N개의 줄에 걸쳐 숫자 카드 묶음의 각각의 크기가 주어짐

**출력 조건**
+ 첫째 줄에 최소 비교 횟수를 출력 (21억 이하)

**입력 예시**  
3  
10  
20  
40  

**출력 예시**  
100

```python
n = int(input())
card = []
for _ in range(n):
    card.append(int(input()))

sorted(card)

count = []
for i in range(1, len(card)):
    if i==1:
        # 제일 처음엔 0번째와 1번째 원소 더해주기
        count.append(card[0]+card[1])
    else:
        # 두개 더한 값에 card[i] 값 더해주기
        count.append(count[i-2] + card[i])
# count : [30,70]
print(sum(count))
```
