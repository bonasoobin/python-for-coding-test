### 퇴사

[백준 링크](https://www.acmicpc.net/problem/14501) []

+ 오늘부터 N+1일째 되는 날 퇴사를 하기 위해서, 남은 N일 동안 최대한 많은 상담을 하려고 함
+ 하루에 하나씩 서로 다른 사람의 상담을 잡아놓음
+ 각각의 상담은 상담을 완료하는 데 걸리는 기간 Ti와 상담을 했을 때 받을 수 있는 금액 Pi로 이루어져 있음

+ N=7일 때, 상담 일정은 다음과 같음

+ 1일에 잡혀 있는 상담은 총 3일이 걸리고, 상담했을 때 받을 수 있는 금액은 10
+ 5일에 잡혀 있는 상담은 총 2일이 걸리며, 받을 수 있는 금액은 15

+ 상담하는 데 필요한 기간이 1일보다 클 수 있기 때문에, 모든 상담을 할 수 없음
+ 예를 들어, 1일에 상담을 하게 되면, 2-3일에 있는 상담은 할 수 없음
+ 2일에 있는 상담을 하게 되면 4-6일에 있는 상담은 할 수 없음
+ N+1일째에는 회사에 없기 때문에, 6-7일에 있는 상담은 할 수 없음

+ 퇴사 전에 할 수 있는 상담의 최대 이익은 1,4,5일에 상담을 하는 것이며 이때의 이익은 45임
  
+ 상담을 적절히 했을 때, 얻을 수 있는 최대 수익 구하는 프로그램 작성하기

**입력 조건**  
+ 첫째 줄에 N(1 <= N <= 15)이 주어짐
+ 둘째 줄부터 N개의 줄에 Ti와 Pi가 공백으로 구분되어서 주어지며, 1일부터 N일까지 순서대로 주어짐  
  (1 <= Ti <= 5, 1 <= Pi <= 1000)

**출력 조건**  
+ 첫째 줄에 얻을 수 있는 최대 이익 출력

```python
n = int(input())
consulting = []
for _ in range(n):
    consulting.append(list(map(int, input().split())))

#1일부터 각 날짜에 시작하기
pay = []
for i in range(n):
    dayy = 0
    payy = 0
    while (dayy <= n):
        if (i+dayy+day) <= n:
            day = consulting[i+dayy][0] # day는 지금 몇번째 인덱스에 있는지 확인
            payy += consulting[i+dayy][1] # payy는 i+1일부터 시작했을때 누적 이익
            dayy += consulting[i+dayy][0] # dayy는 i+1일부터 시작했을때 누적 일수
        else:
            break
    pay.append(payy)
print(max(pay))
```
